```python
#Import all software
import os
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Convert datum column 
pharma_monthly.datum = pd.to_datetime(pharma_monthly.datum).dt.year
```

```python
# Compute mean of total dataset for years (2014-2019)
pharma_monthly.groupby("datum").mean()
```

```python
#Construct histogram chart comparing mean of each category 
colors = plt.cm.tab20(np.linspace(0, 1, len(pharma_monthly.groupby("datum").mean())))


pharma_monthly.groupby("datum").mean().plot(kind="bar",color=colors,figsize=(10,5),width=.6)
plt.title(label=' Average Sales Comparison Per Drug Category (2014-2019)',color ="Black",fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.ylabel('Count ', fontweight = 'bold')


plt.show()
plt.close()
```

```python
# Compute median of total dataset for years (2014-2019)
pharma_monthly.groupby("datum").median()
```

```python
#Construct histogram chart comparing median of each category 
colors = plt.cm.tab20(np.linspace(0, 1, len(pharma_monthly.groupby("datum").median())))


pharma_monthly.groupby("datum").median().plot(kind="bar",color=colors,figsize=(10,5),width=.6)
plt.title(label=' Median Sales Comparison Per Drug Category (2014-2019)',color ="Black",fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.ylabel('Count ', fontweight = 'bold')


plt.show()
plt.close()
```

```python
# Compute standard deviation of total dataset for years (2014-2019)
pharma_monthly.groupby("datum").std()
```

```python
#Construct histogram chart comparing median of each category 
colors = plt.cm.tab20(np.linspace(0, 1, len(pharma_monthly.groupby("datum").std())))


pharma_monthly.groupby("datum").std().plot(kind="bar",color=colors,figsize=(10,5),width=.6)
plt.title(label=' Standard Deviation Sales Comparison Per Drug Category (2014-2019)',color ="Black",fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.ylabel('Count ', fontweight = 'bold')


plt.show()
plt.close()
```
